# Task2:- Image to Pencil Sketch with Python

##### Step-1: Install OpenCV library and then import cv2

```python
import matplotlib.pyplot as plt
import cv2
%matplotlib inline
```

#### Step-2: Read the image using imread function

```python
image = cv2.imread("cat.jpg")
```

```python
plt.imshow(image[:,:,::-1])
plt.axis(False)
plt.show()
```

##### Step3: Convert the image from BGR to RGB

```python
img_RGB = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
```

#### Step-4:Converting the original image to greyscale

```python
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
```

```python
inverted_image = 255 - gray_image
```

#### Step-5:Blur the image using Gaussian function in OpenCV

```python
blurred = cv2.GaussianBlur(inverted_image, (21, 21), 0)
```

#### Step-6:Invert the Blur image

```python
inverted_blurred = 255 - blurred
pencil_sketch = cv2.divide(gray_image, inverted_blurred, scale=256.0)
```

#### Step-7:Ploting original and sketch image

```python
plt.figure(figsize = (16,10))
plt.subplot(1,2,1)
plt.title('Original Image', size = 20)
plt.imshow(img_RGB)
plt.axis('off')

plt.subplot(1,2,2)
plt.title('Pencil Sketch', size = 20)
sketch_rgb = cv2.cvtColor(pencil_sketch, cv2.COLOR_BGR2RGB)
plt.imshow(sketch_rgb)
plt.axis('off')

plt.show()
```
